```python
import pandas as pd
#load dataset
df = pd.read_csv("AfricaCupOfNationsMatches.csv")
print(df.head())  # Shows first few rows to confirm it's loaded
```

```python
print(df.head(7))
#print first 7 rows
```

```python
selected_columns = df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
print(selected_columns)
#select specific columns
```

```python
egypt_matches = df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
print(egypt_matches)
#rows where Egypt appears
```

```python
print("Rows and Columns:", df.shape)
#number of rows and columns
```

```python
missing_attendance = df[df['Attendance'].isna()]
print(missing_attendance)
#rows where attendance is missing
```

```python
filtered_goals = df[df['HomeTeamGoals'].between(3, 6)]
print(filtered_goals)
#where homegoals are between 3 and 6
```

```python
df.at[2, 'AwayTeamGoals'] = 10
print(df.iloc[2])
# change awayteamgaols in 3rd row to 10
```

```python
sorted_df = df.sort_values(by=['HomeTeam', 'HomeTeamGoals'], ascending=[True, False])
print(sorted_df)
#sort by hometeam(ascending)
```

```python
columns_list = df.columns.tolist()
print(columns_list)
#get list of column headers
```

```python
df['TournamentYear'] = 2025  # Example column
print(df.head())
#add new column of your choice
```

```python
new_rows = pd.DataFrame([
    {'HomeTeam': 'Uganda', 'AwayTeam': 'Kenya', 'HomeTeamGoals': 1, 'AwayTeamGoals': 2},
    {'HomeTeam': 'Nigeria', 'AwayTeam': 'Egypt', 'HomeTeamGoals': 3, 'AwayTeamGoals': 3}
])
df = pd.concat([df, new_rows], ignore_index=True)
print(df.tail())
#add two new rows
```

```python
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')
#change uganda in awayteam column
```

```python
df.reset_index(drop=True, inplace=True)
print(df.head())
#reset index of dataframe
```

```python
print('Stadium' in df.columns)
#check if stadium column is present
```

```python
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
#convert int datatype to float
```

```python
df = df[:-10]
#remove last 10 rows
```

```python
for index, row in df.iterrows():
    print(f"{row['HomeTeam']} vs {row['AwayTeam']} – {row['HomeTeamGoals']}:{row['AwayTeamGoals']}")
#iterate over rows
```

```python
df = df[['AwayTeam', 'HomeTeam', 'HomeTeamGoals', 'AwayTeamGoals']]  # Example reorder
print(df.head())
#change order of columns
```
