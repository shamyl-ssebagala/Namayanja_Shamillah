```python
import pandas as pd
```

```python
#load the datasets
df1 = pd.read_csv("Sales.csv")
df2 = pd.read_csv("Mine.csv")
print(df1)
print(df2)
```

```python
#cleaning for Sales dataset
#drop completely empty rows and columns
df1.dropna(how='all', inplace=True)
print(df1)
```

```python
#convert date column to datetime
if 'date' in df1.columns:
    df1['date'] = pd.to_datetime(df1['date'], errors='coerce')
    print(df1)
```

```python
#remove duplicates
df1.drop_duplicates(inplace=True)
print(df1)
```

```python
#cleaning for mine dataset
#drop completely empty rows and columns
df2.dropna(how='all', inplace=True)
print(df2)
```

```python
#convert date column to datetime
if 'date' in df1.columns:
    df2['date'] = pd.to_datetime(df2['date'], errors='coerce')
    print(df2)
```

```python
#remove duplicates
df2.drop_duplicates(inplace=True)
print(df2)
```

```python
# Remove irrelevant columns (example: 'duration')
irrelevant2 = ['duration', 'flag']
df2.drop(columns=[col for col in irrelevant2 if col in df2.columns], inplace=True)
print(df2)
```

```python
#cleaned dataset
df1.to_csv("cleaned_dataset1.csv", index=False)
df2.to_csv("cleaned_dataset2.csv", index=False)
print(df1)
print(df2)
```
